```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
auto_data=pd.read_csv('imports-85.data.txt',header=None)
```

```python
auto_data.head()
```

```python
columns=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style',
        'drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type',
         'num-of-cylinders','engine-size','fuel-system','borke','stroke','compression-ratio','horsepower','peak-rpm',
         'city-mpg','highway-mpg','price'
        ]
```

```python
auto_data.columns=columns
```

```python
auto_data.head()
```

```python
auto_data=auto_data.replace('?',np.nan)
```

```python
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe()
```

```python
auto_data['price']=pd.to_numeric(auto_data['price'],errors='coerce')
```

```python
auto_data['price'].describe()
```

```python
auto_data=auto_data.drop('normalized-losses',axis=1)
```

```python
auto_data.describe()
```

```python
auto_data['horsepower']=pd.to_numeric(auto_data['horsepower'],errors='coerce')
```

```python
auto_data.describe()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
auto_data['num-of-cylinders']
```

```python
cylinders_dict={'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'eight':8,
    'twelve':12
}
auto_data['num-of-cylinders'].replace(cylinders_dict,inplace=True)
auto_data.head()
```

```python
auto_data=pd.get_dummies(auto_data,columns=['make','fuel-type','aspiration','num-of-doors'
                                           ,'body-style','drive-wheels','engine-location','fuel-system','engine-type'])
```

```python
auto_data.head()
```

```python
auto_data=auto_data.dropna()
auto_data
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=auto_data.drop('price',axis=1)
```

```python
y=auto_data['price']
```

```python
print(X.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
linear_model=LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
linear_model.score(x_train,y_train)
```

```python
linear_model.coef_
```

```python
predictors=x_train.columns
coef=pd.Series(linear_model.coef_,predictors).sort_values()
coef
```

```python
y_pred=linear_model.predict(x_test)
```

```python
plt.figure(figsize=(15,6))
plt.plot(y_pred,label='predicted label')
plt.plot(y_test.values,label='actual label')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
r2_score=linear_model.score(x_test,y_test)
```

```python
r2_score
```

```python
from sklearn.metrics import mean_squared_error
linear_model_mse=mean_squared_error(y_pred,y_test)
linear_model_mse
```

```python
import math
math.sqrt(linear_model_mse)
```

# lasso and rigde regression


```python
from sklearn.linear_model import Lasso,Ridge
```

```python
lasso_model=Lasso(alpha=5,normalize=True)
```

```python
lasso_model.fit(x_train,y_train)
```

```python
lasso_model.score(x_train,y_train)
```

```python
coef=pd.Series(lasso_model.coef_,predictors).sort_values()
coef
```

```python
y_predict=lasso_model.predict(x_test)
```

```python
plt.figure(figsize=(15,6))
plt.plot(y_predict,label='predicted label')
plt.plot(y_test.values,label='actual label')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
lasso_model.score(x_test,y_test)
```

```python
lasso_mse=mean_squared_error(y_predict,y_test)
lasso_mse
```

```python
math.sqrt(lasso_mse)
```

# Ridge Regression

```python
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_train,y_train)
```

```python
coef=pd.Series(ridge_model.coef_,predictors).sort_values()
coef
```

```python
y_predicts=ridge_model.predict(x_test)
```

```python
ridge_model.score(x_test,y_test)
```

```python
ridge_mse=mean_squared_error(y_test,y_predicts)
ridge_mse
```

```python
math.sqrt(ridge_mse)
```
