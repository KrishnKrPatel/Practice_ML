```python
#importing some useful python libraies and package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
import seaborn as sn
```

```python
#importing the dataset 
dataset=pd.read_csv('Salary_Data.csv')
#spliting the dataset into the features and labels.
dataset.head()
X=dataset.drop('Salary',axis=1)
Y=dataset.drop('YearsExperience',axis=1)
#ploting the features vs label
plt.plot(X,Y,color="blue")
plt.xlabel('YearsExperience',color='red')
plt.ylabel('Salary',color='green')
plt.show()
```

```python
#ploting the features vs label using scatter plot
plt.scatter(X,Y,color="blue")
plt.xlabel('YearsExperience',color='red')
plt.ylabel('Salary',color='green')
plt.show()
```

```python
#spliting the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=1/4.,random_state=0)
#creating the objects for the linearRegression models
lm=LinearRegression()
lm.fit(X_train,y_train)
y_train_predict=lm.predict(X_train)
y_test_predict=lm.predict(X_test)
#ploting the best line which are best fitted on the training data
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,y_train_predict,color='purple')
plt.title('Salary Vs Experience(Training_set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#ploting the best line which are best fitted on the test data
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,y_test_predict,color='purple')
plt.title('Salary Vs Experience(Training_set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#calculating the mean_squared_error
from sklearn.metrics import mean_squared_error
X_train_mse=np.mean((y_train-y_train_predict)**2.0)
X_test_mse=mean_squared_error(y_test,y_test_predict)
print("The X_train_mse:{}".format(X_train_mse))
print("The X_test_mse:{}".format(X_test_mse))
print(lm.coef_[0][0])
print(lm.intercept_[0])
print("The line which is best fitted on the data:{}*x+{}".format(lm.coef_[0][0],lm.intercept_[0]))
```

```python
#using the ridge algorithms
lm=Ridge(alpha=0.5)
lm.fit(X_train,y_train)
y_train_predict=lm.predict(X_train)
y_test_predict=lm.predict(X_test)
#ploting the best line which are best fitted on the data
plt.scatter(X,Y)
plt.plot(X_train,y_train_predict,color='purple')
plt.show()
from sklearn.metrics import mean_squared_error
X_train_mse=np.mean((y_train-y_train_predict)**2.0)
X_test_mse=mean_squared_error(y_test,y_test_predict)
print("The X_train_mse:{}".format(X_train_mse))
print("The X_test_mse:{}".format(X_test_mse))
```
